# Read results

```python
import os
import pickle

import pandas as pd
import numpy as np
```

## Leukemia

```python
p_values = [2000, 4000, 6000, 8000, 10000]
n__values =  [50, 40, 30] # not the real n's (which are computed in the notebook)s!
percent_values = [1.25, 2.5, 5.0, 10.0]
```

```python
cwd = os.getcwd()
cwd
```

```python
p = 2000

results_folder = os.path.join(cwd, f"results/synthetic_data_{p}")
results_folder
```

```python
results_d = {}

for p in p_values:

    results_folder = os.path.join(cwd, f"results/synthetic_data_{p}")
    results_folder

    ns = [int(p//n_) for n_ in n__values]
    nzs = [int(p*percent//100) for percent in percent_values]

    fts_modes = ["full", "random", "k-best", "k-best-mi", "pk-lpnn"]
    models = ["knn", "lr", "nb-gaussian", "svc"]
        
    results_d[p] = {}

    for n in ns:

        results_d[p][n] = {}

        for nz in nzs:

            results_d[p][n][nz] = {}

            for fts_mode in fts_modes:

                results_d[p][n][nz][fts_mode] = {}

                for model in models:

                    with open( os.path.join(results_folder, f"{n}", f"{nz}", f"{fts_mode}_{nz}_{model}.pkl"), "rb" ) as fh:

                        results = pickle.load(fh)
                        f1 = np.mean(results["f1"]).round(3)
                        b_acc = np.mean(results["b_acc"]).round(3)
                        #Â print("f1: ", f1, "b_acc: ", b_acc)

                        results_d[p][n][nz][fts_mode][model] = (f1, b_acc)
```

```python
results_df = {}

for p in p_values:
    
    print("****************")
    print(f"*** p = {p} ***")
    print("****************\n")
    
    ns = [int(p//n_) for n_ in n__values]
    nzs = [int(p*percent//100) for percent in percent_values]
    
    results_df[p] = {}
    
    for n in ns:

        results_df[p][n] = {}

        for nz in nzs:

            results_df[p][n][nz] = pd.DataFrame.from_dict(results_d[p][n][nz])
            
            print(f"*** n = {n}, N_z = {nz} ***")
            display(results_df[p][n][nz])
            print("\n")
```

**Remarks**

- For sparse-pca, the computation time increases with the number of selected features.
- For Pk-LPNN, this is not the case: computation time independent from nb of selected features.
- Pk-LPNN seems more robust: constant solid results across datasets, as opposed to sparse-pca (good and bad).
- Pk-LPNN beats k-best in some cases.
- **For the Prostate dataset, Pk-LPNN is the best method!**
- For other datasets, no clear pattern.
- Our approach open the research avenue for using compressive sampling methods for feature selection.

## DLBCL: tables

```python
results_folder = os.path.join("../results/DLBCL")
results_folder
```

```python
nzs = ["68", "137", "273", "547"]
fts_modes = ["full", "random", "k-best", "k-best-mi", "sparse-pca", "lasso", "pk-lpnn"]
models = ["knn", "lr", "nb-gaussian", "svc"]

results_d = {}

for nz in nzs:

    results_d[nz] = {}
    
    for fts_mode in fts_modes:
    
        results_d[nz][fts_mode] = {}
        
        for model in models:
            
            with open( os.path.join(results_folder, f"{nz}", f"{fts_mode}_{nz}_{model}.pkl"), "rb" ) as fh:
                
                results = pickle.load(fh)
                f1 = np.mean(results["f1"]).round(3)
                b_acc = np.mean(results["b_acc"]).round(3)
                # print("f1: ", f1, "b_acc: ", b_acc)
    
                results_d[nz][fts_mode][model] = (f1, b_acc)
```

```python
results_df = {}

for nz in nzs:
    
    results_df[nz] = pd.DataFrame.from_dict(results_d[nz])
```

## DLBCL: plots

```python
# results_folder = os.path.join("C:/Users/DELL/Documents/Python/Application/Fer application/genetic_data/genetic_data_DLBCL") # XXX
results_folder = os.path.join("../results/DLBCL")
results_folder
```

```python
nzs = ["68", "137", "273", "547"]
fts_modes = ["full", "random", "k-best", "k-best-mi", "sparse-pca", "lasso", "pk-lpnn"]
models = ["knn", "lr", "nb-gaussian"] #"svc"]
max_nz = 5469  # Valeur maximale de Nz

# Calcul des pourcentages
nzs_percent = [int(nz) / max_nz * 100 for nz in nzs]
```

### Classifiers

```python
### F1 score
```

```python
# Stockage des résultats
results_d = {}

for fts_mode in fts_modes:
    results_d[fts_mode] = {model: [] for model in models}
    
    for nz in nzs:
        for model in models:
            # Construire le chemin vers le fichier
            file_path = os.path.join(results_folder, f"{nz}", f"{fts_mode}_{nz}_{model}.pkl")
            
            # Vérification et lecture
            if os.path.exists(file_path):
                with open(file_path, "rb") as fh:
                    results = pickle.load(fh)
                    f1 = np.mean(results["f1"]).round(3)  # Moyenne de F1
                    results_d[fts_mode][model].append(f1)
            else:
                print(f"File not found: {file_path}")
                results_d[fts_mode][model].append(None)  # Aucun résultat si le fichier est absent
```

```python
# save_folder = os.path.join("C:/Users/DELL/Documents/Python/Application/Fer application/genetic_data/genetic_data_DLBCL/PLOT") # XXX
save_folder = os.path.join("../results/DLBCL/plots")
os.makedirs(save_folder, exist_ok=True)
```

```python
# Tracer et sauvegarder les résultats
for fts_mode, models_data in results_d.items():
    if fts_mode == "pk-lpnn":
        fts_mode = "LPNN-FS" 
    elif fts_mode == "sparse-pca":
        fts_mode = "sparse PCA"
    elif fts_mode == "lasso":
        fts_mode = "LASSO"
    elif fts_mode == "k-best-mi":
        fts_mode = "k-best (MI)"
    
    plt.figure(figsize=(7, 3))  # Créer une nouvelle figure pour chaque méthode de sélection de caractéristiques
    for model, scores in models_data.items():
        if model == "knn":
            model = "KNN"
        elif model == "lr":
            model = "LR"
        if model == "nb-gaussian":
            model = "NBC"
        plt.plot(nzs_percent, scores, label=model, linewidth=2, marker='o')  # Courbe pour chaque classifieur
    
    # Titre avec le nom de données (Leukemia) ajouté
    # plt.title(f"F1 Score for DLBCL pour la méthode {fts_mode}")
    plt.xlabel("Nz (%)", size=14)
    plt.xticks(size=14)
    plt.ylabel("F1 Score", size=14)
    plt.yticks(size=14)
    plt.legend(loc="center right")
    plt.grid(False)
    
    # Sauvegarde dans un fichier PDF
    save_path = os.path.join(save_folder, f"F1_{fts_mode}_DLBCL_classifiers.pdf")
    plt.savefig(save_path)
    plt.show()
    plt.close()  # Fermer la figure pour éviter de garder en mémoire

    print(f"Plot saved in: {save_path}")
```

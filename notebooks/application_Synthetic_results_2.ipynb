# Read results

```python
import os
import pickle

import pandas as pd
import numpy as np
```

## Synthetic data

```python
p_values = [1000, 2000, 3000, 4000, 5000]
n__values =  [10] # not the real n's (which are computed in the notebook)!
percent_relevent_values = [5.0]
percent_values = [2.5, 3.75, 5.0, 6.25, 7.5]
```

```python
cwd = os.getcwd()
cwd
```

```python
results_d = {}

# level 1
for p in p_values:

    results_folder = os.path.join(cwd, f"results/synthetic_data_2_{p}")
    results_folder

    ns = [int(p//n_) for n_ in n__values]
    nzs = [int(p*percent//100) for percent in percent_values]
    nrs = [int(p*percent//100) for percent in percent_relevent_values]

    fts_modes = ["full", "random", "k-best", "k-best-mi", "lasso", "pk-lpnn"]
    models = ["knn", "lr", "nb-gaussian", "svc"]
        
    results_d[p] = {}
    
    # level 2
    for n in ns:
        
        results_d[p][n] = {}
        nr = nrs[0]  # XXX
        
        # level 3
        for nz in nzs:

            results_d[p][n][nz] = {}
            
            for fts_mode in fts_modes:

                results_d[p][n][nz][fts_mode] = {}

                for model in models:

                    with open( os.path.join(results_folder, f"{n}_{nr}", f"{nz}", f"{fts_mode}_{nz}_{model}.pkl"), "rb" ) as fh:

                        results = pickle.load(fh)
                        f1 = np.mean(results["f1"]).round(3)
                        b_acc = np.mean(results["b_acc"]).round(3)
                        # print("f1: ", f1, "b_acc: ", b_acc)

                        results_d[p][n][nz][fts_mode][model] = (f1, b_acc)
```

```python
results_df = {}

for p in p_values:
    
    print("****************")
    print(f"*** p = {p} ***")
    print("****************\n")
    
    ns = [int(p//n_) for n_ in n__values]
    nzs = [int(p*percent//100) for percent in percent_values]
    nrs = [int(p*percent//100) for percent in percent_relevent_values]
    
    results_df[p] = {}
    
    for n in ns:
        
        results_df[p][n] = {}
        nr = nrs[0]  # XXX

        for nz in nzs:

            results_df[p][n][nz] = pd.DataFrame.from_dict(results_d[p][n][nz], 
                                                          orient='index')     # XXX if changed here, then changees column_format and header
            
            print(f"*** n = {n}, nb_relevant = {nr} ({n}_{nr}), N_z = {nz} ***")
            display(results_df[p][n][nz])
            print(results_df[p][n][nz].to_latex(bold_rows=True, 
                                                column_format="l | cccc", 
                                                header=["K-NN","LR","NBC","SVC"])
                 ) # latex code
            print("\n")
```
